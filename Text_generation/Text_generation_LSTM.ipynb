```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,LSTM
import np_utils  # numpy related utilities
import numpy as np
# import keras.utils.to_categorical
from tensorflow.keras.utils import to_categorical
```

```python
text = '''In the realm of artificial intelligence, advancements in natural language processing and machine learning have ushered in a new era of possibilities. These technologies, fueled by large-scale neural network models, enable computers to understand and generate human-like text, translating to applications ranging from chatbots that engage in meaningful conversations to language models that can assist in creative writing. As AI continues to evolve, the ethical considerations surrounding its use, data privacy concerns, and the quest for responsible and transparent AI development become increasingly critical. Balancing innovation with ethical and societal implications is a key challenge in shaping the future landscape of artificial intelligence.'''
```

```python
# mapping charecters with integers
unique_char = sorted(list(set(text)))
char_to_int = {} # {'a':0,'b':1}
int_to_char = {} # {0:'a',1:'b'}
for i,c in enumerate(unique_char):  
  char_to_int.update({c:i})  #'\n':0
  int_to_char.update({i:c})  #0:'\n'
```

```python
char_to_int
```

```python
int_to_char
```

```python
x = []
y = []
for i in range(0,len(text)-50):
  sequence = text[i : i+50]  #text[0 to 49]
  label = text[i+50]  # text[50]
  x.append([char_to_int[char] for char in sequence])
  y.append(char_to_int[label])
```

```python
# reshaping of x , normalising and one hot encoding
x_modified = np.reshape(x,(len(x),50,1))
 #reshaping -(samples,time steps,features)
x_modified = x_modified / len(unique_char)  #normalizing
y_modified = to_categorical(y)  #one hot encoding
print(x_modified.shape)
print(y_modified.shape)
```

```python
x_modified
```

```python
y_modified
```

```python
model = Sequential()
model.add(LSTM(300, input_shape=(x_modified.shape[1], x_modified.shape[2]),
               return_sequences=True))
 #return sequence=return the full sequence of outputs for each input sequence
model.add(Dropout(0.2))
model.add(LSTM(800))
model.add(Dropout(0.1))
model.add(Dense(y_modified.shape[1],activation='softmax'))
# y modified.shape[1]=number of unique classes or categories in your classification task.
model.compile(loss='categorical_crossentropy',optimizer='adam')
```

```python
# Fitting the model
model.fit(x_modified,y_modified,epochs=50 , batch_size=30)
```

```python
# picking random seed
start_index = np.random.randint(0,len(x)-1)
new_string = x[start_index]  # x is a list of lists
# prediction
new_string_reshaped = np.reshape(new_string,(1,len(new_string),1))
new_string_reshaped = new_string_reshaped/len(unique_char)
pred_list = model.predict(new_string_reshaped)
# generates a list of probabilities for each char
pred = np.argmax(pred_list)
```

```python
pred
```

```python
# printing input sequence and predicted character
character_out = int_to_char[pred]
character_in = []
for value in new_string:
  character_in.append(int_to_char[value])
```
